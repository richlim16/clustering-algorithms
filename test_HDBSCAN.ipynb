```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import hdbscan

%matplotlib inline
sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds = {'alpha': 0.5, 's': 80, 'linewidths': 0}
```

```python
test_data = pd.read_csv('Mall_customers.csv', encoding='Latin-1') 
test_data.head()
```

```python
test_data = test_data.iloc[:,[3,4]].values
```

```python
test_data
```

```python
plt.scatter(test_data[:,0], test_data[:,1], color='b', **plot_kwds)
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=9, gen_min_span_tree=True)
clusterer.fit(test_data)
```

```python
clusterer.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
clusterer.condensed_tree_.plot()
```

```python
clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
```

```python
palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.5, 0.5, 0.5) for col, sat in
                  zip(clusterer.labels_, clusterer.probabilities_)]
plt.scatter(test_data.T[0], test_data.T[1], c=cluster_colors, **plot_kwds)
plt.title("HDBSCAN Clustering")
plt.show()
```

The mutual reachability distance is across a wider range
and the condensed tree plot has a wider half life value(?)
as the min_cluster_size increases
